```python
# Plotting 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
SIMD = pd.read_csv('outputFiles/SIMD.txt', sep=" ", header=None)
CUDA = pd.read_csv('outputFiles/CUDA.txt', sep=" ", header=None)
OMP = pd.read_csv('outputFiles/OMP.txt', sep=" ", header=None)
```

```python
f = plt.figure()
plt.scatter(SIMD[0],SIMD[1], c='b', label='SIMD', marker='o',alpha=.5, s=10)
plt.scatter(CUDA[0],CUDA[1], c='r', label='CUDA', marker='1',alpha=.5,s=10)
plt.scatter(OMP[0],OMP[1], c='y', label='OMP', marker='x',alpha=.5,s=10)
plt.legend(loc='upper left')
plt.title('SIMD, CUDA, OMP Auto Correlation VS Shift')
plt.xlabel('Shift')
plt.ylabel('Auto Correlation Sums')
plt.show()
f.savefig('plot.pdf')
```
